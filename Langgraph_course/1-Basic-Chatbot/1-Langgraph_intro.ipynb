## What is Langgraph

- LangGraph is an orchestration framework for buidling intelligent, stateful and multi-step LLM workflows
- It enables advanced features like parallelism, loops, branching, memory, and resumability - making it ideal for agentic and production grade AI applications

- It models your logic as a graph of nodes(tasks) and edges(routing) instead of a linear chain.

### LLM Workflows

1. LLM workflows are step by step process using which we can build complex LLM applications.
2. Each step in a workflow performs a distinct task - such as prompting, reasoning, tool calling, memory access, or decision making.
3. Workflows can be linear, parallel, branched, or looped, allowing for complex behaviours like retries, multi agent communication or tool augmented reasoning
4. Common Workflows

### Example of LLM workflows

1. **Prompt Chaining** - Multiple LLM calls made linearly
2. **Routing** - Decide which LLM will answer what query
3. **Parallelization** - Multiple LLM calls running sub tasks parallely
4. **Orchestrator Workers** - An orchestrator agent decides which subagent gets what task to perform, but they all still work parallely
5. **Evaluator Optimizer** - A generator- evaluator loop to evaluate LLM responses to produce better results through iterative improvement loop.

### Graphs, Nodes and Edges

### Nodes : 

Every node in a langgraph graph is a task that needs to be performed and it is essentially a python function. So we create python functions and they act as our nodes 

### Edges : 
They tell us which node to execute after the completion of the execution of the previous node.

These can be of many types:
- Sequential
- Parallel
- Conditional
- Loop 

### State : Very Important

In Langgraph state is a shared memory that flows through your workflow - it holds all the data being passed between nodes as your graph runs.

You are required to define the state before the formation of the graph in langgraph 

Each node has access to this state before its execution and makes changes to this after its execution

### Reducers

Reducers in Langgraph define how updates from the nodes are applied to the shared state.

Each key in the state can have its own reducer, which determines whether new data replaces, merges or adds to the exisiting value.
