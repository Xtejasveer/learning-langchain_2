## Streaming and Batch

### Streaming
- Most models can stream their output content while it is being generated. By displaying the output progressively, streaming significantly improves user experience, particularly for longer responses

```python
model.invoke("Write me a 200 words paragraph on Aritificial Intelligence")
```

```python
model.stream("Write me a 200 word paragraph on Artificial Intelligence")
```

```python
for chunk in model.stream("Write me a 200 word paragraph on Artificial Intelligence"):
    print(chunk.text, end = "|", flush =True)
```

```python
for chunk in model.stream("Why do parrots have colorful feathers?"):
    print(chunk.text, end = "|", flush =True)
```

### batch
- Batching a collection of independent requests to a model can significantly improve performance and reduce costs, as the processing can be done in parallel:

```python
responses = model.batch([
    "Why do parrots have colored feathers?",
    "How do airplanes fly ?",
    "what is quantum computing?"
])
for response in responses:
    print(response.content)
```
